It looks slicing work for dataframe



*   indexing in dataframe looks for explicit indices
*   slicing, however checked for implicit indices



This can be a cause for confusion



*   To avoid this pandas special indexers loc and iloc
*   We will look at these in a bit lets look at them one by one



1. loc -> provide indexing and slcing that always refrences the explicit index

```python
df.loc[1]
```

```python
df.loc[1:3]
```

```python
# did you noticed something string here

# The range id inclusive of end pioint for loc
# Row with label 3 is included in the result
```

2. iloc -> Allow indexing and slicing that always references the implicit python index

```python
df.iloc[1]
```

```python
df.iloc[0:2]
```

```python
# Now, will iloc consider the range inclusive? -> No, because iloc works with python indices
```

Which on should we use?



*   Generally explicit indexing is considered to be better then explicit
*   But it is recommended to always use both loc and iloc to avoid any confusion



```python
# Now, will iloc consider the range inclusive? -> No, because iloc works with python indices
```

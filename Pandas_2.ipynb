Cancatenating DataFrames

Let's use a mini-case, of users and messages

```python
#users -> stores the user details - IDs and Names of users
```

```python
users = pd.DataFrame({
    'user_id': [1,2,3],
    'name': ['shaadha','shahid','khusali']})
```

```python
users
```

```python
msgs=pd.DataFrame({
    'user_id':[1,2,3,4],
    'message':['hmm','acha','theek hai','nice']})
```

```python
msgs
```

```python
pd.concat([users,msgs])
```

How exactly did concat work?



*   By default, axis=0 (row-wise) for cancatination
*   userid, being same in both dataframes, was combined into a single column.


*   First value of users dataframe were placed, with vlaue of column msg and NaN
*  Then values of msgs dataframe were placed with values of column msg and NaN

-> The Original indices of the rows were preserved.





Now how can we make the indices unique for each row?

```python
pd.concat([users,msgs],ignore_index=True)
```

How can we concatinate them horizontaly?

```python
pd.concat([users,msgs],axis=1)
```

As you can see here



*   Both the dataframes are combined horizontally (column-wise)
*   It gives 2 columns with different positinal (implicit) index, but same label



Mergeing Dataframes

```python
# So far we have only concatenated and not merged data

# But what is the difference between concat and merge?
```

```python
# -> Concat : simply stacks multiple Dataframes together along an axis
# -> Merge : Combine data frame in a smart way based on value in shared columns
```

How can we know tha name of person who send the msg?

```python
# We need information from both the dataframes
# So, can we use pd.concat() for comnining the dataframes?
# Ans-> No
```

```python
pd.concat([users,msgs],axis=1)
```

What are the problems with concat here?



*   concat simply combined/cracked the dataframe horizontally
*   If you notice, userid 3 for user dataframe is stacked against userid 2 for msg dataframe.


*   This way of stacking doesn't help us gain any insights





```python
# ==> pd.concat() does not work according to the values in the columns

# We need to merge the data
```

```python
users.merge(msgs,on='user_id')
```

```python
# only the userid common in both dataframes in shown
```

What type of join is this?

```python
# 1. inner join

# -> Remember joins in SQL

# Inner join, Outer join, left join, right join
```

```python
users.merge(msgs,on='user_id',how='inner')  # Here we specified key similer to promary key in sql
```

```python
users.merge(msgs,on='user_id',how='outer') # All mising value will be replced by NaN
```

```python
users.merge(msgs,on='user_id',how='left')  # This will give info only for the users who sent a message
```

```python
users.merge(msgs,on='user_id',how='right')
```

```python
# Note -> NaN in name can be thought of as an anonymous message

# But sometimes the column names might be different even if they contain the same data

# Let's rename our users column userid to id
```

```python
users.rename({'user_id':'id'},inplace=True,axis=1)
```

```python
users
```

Now, how can we merge the 2 dataframes when the key has different name?

```python
users.merge(msgs,left_on='id',right_on='user_id')
```
